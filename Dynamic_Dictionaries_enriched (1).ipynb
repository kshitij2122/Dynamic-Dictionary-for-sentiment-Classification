```
def decontracted(phrase):
    # specific
    phrase = re.sub(r"won\'t", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)

    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase
```

```
#Username and site link removal from tweet
import re
df['Tweet']=df['Tweet'].apply(lambda x:re.sub('@[\w]+','',x))
df['Tweet']=df['Tweet'].apply(lambda x:re.sub(r"http\S+", "", x))
df['Tweet']=df['Tweet'].apply(lambda x:decontracted(x))
```

# **Function 1**: To initially classify the sentences into Negative(0), Positive(1) & Unidentified(2)

```
#JUDGING SENTENCES FOR POSITIVE, NEGATIVE AND UNIDENTIFIED
def judge(sen):
  l=len(sen.split())
  if l==0:
    return 2
  p=0
  n=0
  u=0
  for i in sen.split():
    if i in Pos_list:
      p+=1
    elif i in Neg_list:
      n+=1
    else:
      u+=1
    if u/l>0.5:
      return 2                         #  2 for unidentified sentence         
    elif (n+p)==0:
      return 2                        
    elif p/(n+p)>0.6:
      return 1                         # 1 for positive sentence                      
    elif n/(n+p)>0.6:
      return 0                         #  0 for negative sentence
    else:
      return 2
```

```
a=[]
for i in corpus:
  a.append(judge(i))
import numpy as np
x = np.array(a)
unique, counts = np.unique(x, return_counts=True)
print(np.asarray((unique, counts)).T)
```

```
init_pos={}
init_neg={}
unidentified_corpus=[]
```

# **Function 2**: To add words of the identified positive and negative sentences into two new dictionaries with words as ‘keys’ and their frequencies as ‘values’(init_pos & init_neg).


```
def init_update_dictionary(sen):
  r=judge(sen)
  if r==2:
    unidentified_corpus.append(sen)
  a=sen.split()
  if r==1:                                        
    for i in a: 
      if i in Pos_list:                            #if word exists in the Pos_list         
        if i in init_pos:                               #if word already exists in the dictionary, 
            init_pos[i]+=1                               #increment its frequency             
        else:                                                    
             init_pos[i]=1                                #else add the new word         
  if r==0:                                                        
    for i in a: 
      if i in Neg_list:                                            
         if i in init_neg:                                           
             init_neg[i]+=1
         else:
             init_neg[i]=1
```

```
for i in corpus:
  init_update_dictionary(i)
```

# Function 3: To enrich the positive and negative dictionaries created in Function 2 with all the new words occurring in the positive and negative sentences respectively.


```
def valence(sen):
  sum_pos=0
  sum_neg=0
  for key, value in init_pos.items(): 
    sum_pos=sum_pos+value                                     #Total sum of frequencies of all words in the Positive dictionary

  for key, value in init_neg.items():                         #Total sum of frequencies of all words in the Negative dictionary
    sum_neg=sum_neg+value
  p=0
  n=0
  pos_percentage=0
  neg_percentage=0
  pos_valence=0
  neg_valence=0
  sum_of_valence_of_pos_words=0
  sum_of_valence_of_neg_words=0

  
  for i in sen.split():
    if i in init_pos:
      sum_of_valence_of_pos_words+=(init_pos[i]/sum_pos)*1000       #Sum of valence of Positive words 
    if i in init_neg:
      sum_of_valence_of_neg_words+=(init_neg[i]/sum_neg)*1000       #Sum of valence of Negative words
  
  pos_valence=sum_of_valence_of_pos_words
  neg_valence=sum_of_valence_of_neg_words

  if pos_valence+neg_valence==0:
    return "Unidentified"
  pos_percentage=float(pos_valence)/float(pos_valence+neg_valence)   #Positive percentage of the comment
  neg_percentage=float(neg_valence)/float(pos_valence+neg_valence)   #Negative percentage of the comment
  
  #Defining polarity & Adding new words in the dictionaries
  if pos_valence>neg_valence:
    polarity=1
  elif neg_valence>pos_valence:
    polarity=0   
  if polarity==0:
    for j in i.split():
      if j in init_neg:
        init_neg[j]+=neg_percentage                    
      elif j not in init_neg:
        init_neg[j]=neg_percentage

  elif polarity==1:
    for j in i.split():
      if j in init_pos:
        init_pos[j]+=pos_percentage
      elif j not in init_pos:
        init_pos[j]=pos_percentage      
  if(polarity==1 or polarity==0):
       return polarity, pos_valence, neg_valence
  else:  
       return ("Unidentified")  

```
